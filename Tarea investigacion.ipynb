#**HISTORIGRAMA**


¿QUE ES?

Un histograma es tipo de diagrama estadístico en el que se representa un conjunto de datos estadísticos mediante barras rectangulares, de manera que cada barra del histograma es proporcional a su frecuencia correspondiente.
Los histogramas sirven para representar gráficamente variables continuas, como por ejemplo el peso de una muestra estadística. Además, un histograma permite visualizar rápidamente la forma que tiene una distribución.

¿COMO SE CREAN?

Los pasos para hacer un histograma son:


*Dividir el eje horizontal del histograma en intervalos según la serie de datos.




---



*Representar los valores de las frecuencias de los intervalos en el eje vertical del histograma.



---



*Para cada intervalo, dibujar una barra rectangular con una altura equivalente a la frecuencia del intervalo. Ten en cuenta que las barras de dos intervalos consecutivos deben tocarse.

¿PARA QUE NOS SIRVEN?

Visualizar la distribución de datos : Muestran cómo se distribuyen los valores de una variable continua o discreta agrupándolos en intervalos (o "bins"). Esto permite identificar patrones, como si los datos están distribuidos de forma normal, sesgada o uniforme.

Detectar tendencias y variabilidad : Ayudan a comprender la concentración de datos en ciertos rangos, su dispersión y la presencia de valores atípicos.

Comparar distribuciones : Facilitan la comparación de conjuntos de datos al graficarlos uno junto al otro.

Analizar frecuencia : Muestran con qué frecuencia ocurren valores en un rango específico, lo que es útil en estadísticas descriptivas.

Tomar decisiones basadas en datos : Son una herramienta clave en áreas como la calidad del producto, donde se evalúan variaciones para asegurar estándares.

¿EN QUE CONTEXTO SE UTILIZAN?

Estadísticas y análisis de datos

Análisis de distribución : Comprender cómo se distribuyen los datos, por ejemplo, las calificaciones de los estudiantes en un examen.

Desviación estándar y varianza : Visualizar la dispersión de los datos para identificar si los valores están agrupados o dispersos.

---
Ciencias e investigación

Estudios demográficos : Evaluar la distribución de edades, ingresos o cualquier otra variable cuantitativa en una población.


Mediciones experimentales : Comparar los resultados de experimentos científicos para verificar la consistencia o detectar anomalías.


---

Negocios y economía

Análisis de ventas : Evaluar la frecuencia de ventas en rangos específicos (ejemplo: ventas entre $0 y $50, $50 y $100, etc.).

Calidad y control : Identificar la variabilidad en productos, como el peso de envases o la duración de un componente.


#**DIAGRAMA DE TALLO Y HOJA**

¿QUE ES?



El diagrama de tallo y hoja es un tipo de diagrama estadístico en el que se representa un conjunto de datos cuantitativos.

En un diagrama de tallo y hoja cada dato está separado por la hoja, que es su último dígito, y el tallo, que son las cifras restantes. De modo que en un diagrama de tallo y hoja se pone cada hoja en la línea de su tallo correspondiente.

De esta forma, los diagramas de tallos y hojas permiten representar gráficamente un conjunto de datos estadísticos y, además, ayudan a visualizar la forma de la distribución.

¿COMO SE CREA?

Para hacer un diagrama de tallo y hoja se deben seguir los siguientes pasos:

*Ordenar los datos de menor a mayor.

*Si es necesario, redondear los números hasta que tengan el número de cifras deseadas.

*Decidir en qué cifra se hará el corte entre el tallo y la hoja, es decir, determinar cuántas cifras serán las hojas. Normalmente, se hace que la hoja sea solamente el último dígito de cada dato.

*Representar cada dato en el diagrama de tallo y hoja. Los tallos se colocan en la columna de la izquierda en orden ascendente, mientras que las hojas se ponen en la columna de la derecha a la altura de su tallo correspondiente y también en orden ascendente.

¿PARA QUE NOS SIRVE?



Facilitar el análisis descriptivo :

Calcula fácilmente medidas como la mediana, la moda o el rango sin necesidad de cálculos complejos.
Ayuda a estimar rápidamente la cantidad de datos dentro de ciertos rangos.

Comparar conjuntos de datos:

Los diagramas dobles de tallo y hoja permiten contrastar dos grupos (por ejemplo, calificaciones de dos clases) para ver diferencias en las distribuciones.

#**DIAGRAMA DE CAJA**

¿QUE ES?

El diagrama de caja y bigotes, también llamado diagrama de caja o boxplot, es un gráfico que representa un conjunto de datos estadísticos de manera visual utilizando los cuartiles.

La principal característica del diagrama de caja y bigotes es que permite visualizar rápidamente la dispersión de una serie de datos, ya que indica los cuartiles, la mediana, los valores extremos y los valores atípicos de los datos.

¿COMO SE CREA?

*Ordenar los datos de la muestra estadística.

*Calcular los cuartiles (Q1, Q2 y Q3) y representarlos como la caja del diagrama. El primer y el tercer cuartil corresponden a los límites de la caja, y para representar la mediana (el segundo cuartil) se debe dibujar una línea dentro de la caja donde se encuentre su valor.

*Calcular el rango intercuartílico, que es igual al tercer cuartil menos el primer cuartil.

*Identificar los valores atípicos de la muestra, que son aquellos valores menores que LI o mayores que LS. Representar dichos valores fuera del alcance de los bigotes con puntos.

*Identificar y representar los valores extremos, que és el valor más pequeño y el valor más grande dentro del intervalo formado por LI y LS. Estos valores son el final de los dos bigotes del diagrama.

¿PARA QUE NOS SIRVE?

Analizar la distribución de datos:

Representan de manera clara los rangos principales (mínimo, máximo, cuartiles y mediana) y permiten ver si los datos están sesgados hacia valores altos o bajos.


---


Comparar múltiples conjuntos de datos :

Son ideales para comparar la distribución de varios grupos de datos al colocarlos lado a lado. Por ejemplo, comparar los resultados de diferentes experimentos o grupos de personas.


---


Detectar valores atípicos:

Señalan posibles valores extremos que pueden influir en el análisis y que podrían requerir una investigación más detallada.


---


Resumir información compleja:

Proporcionan una descripción gráfica resumida sin necesidad de observar cada dato individualmente.

¿EN QUE CONTEXTO SE UTILIZAN?

Estadística y análisis de datos:

Comparar el rendimiento de diferentes grupos.
Analizar resultados de encuestas, pruebas o mediciones.


---

Negocios:

Analizar datos financieros, como la variabilidad de ingresos o precios en diferentes mercados.


---
Control de calidad:

Evaluar la consistencia de procesos de manufactura detectando anomalías.
